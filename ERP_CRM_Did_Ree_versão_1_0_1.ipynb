Nome sugerido para a saída do código acima: "Cury Properties Details" CPD

```python

import requests

# Consulta 1: Obter informações sobre imóveis da Cury
url = "https://cury.net/api/products"
params = {
    "adjust_price": 1,
    "bedrooms": "0,1,2,3",
    "limit": 100,
    "page": 1,
    "pricerange": "100,800",
    "regions": "56,47,67,18,45,66,42,43,64,44",
    "state": "sp",
    "status": "5,2,3,4",
    "type": "1,2"
}
response = requests.get(url, params=params)
data = response.json()

if data["success"]:
    print("Consulta 1: Base de Imóveis Cury")
    print("Total de imóveis encontrados:", data["total"])
    print("Lista de imóveis:")
    for item in data["data"]:
        print("ID:", item["id"])
        print("Estado:", item["state"])
        print("Link:", "https://cury.net/" + item["link"])
        print("Título:", item["title"])
        print("Vendido:", item["sold"])
        print("Status:", item["status"]["label"])
        print("Região:", item["region"])
        print("Bairro/Sub-região:", item["nbhood_subregion"])
        print("Capa:", "https://cury.net/" + item["cover"])
        print("Detalhes:")
        for detail in item["details"]:
            print("  -", detail["label"])
        print("Texto de destaque:", item["highlight_text"])
        print("Preço oculto:", item["hide_price"])
        print("\n")

# Consulta 2: Obter link do vídeo do YouTube
url = "https://cury.net/imovel/SP/zona-sul/guedala-park-premium"
response = requests.get(url)
html = response.text

start_index = html.find('src="https://www.youtube.com/embed/') + 5
end_index = html.find('?rel=0&showinfo=0&origin=https://cury.net"', start_index)
youtube_link = html[start_index:end_index]
print("Consulta 2: Link do vídeo do YouTube")
print("Link:", youtube_link)
```
